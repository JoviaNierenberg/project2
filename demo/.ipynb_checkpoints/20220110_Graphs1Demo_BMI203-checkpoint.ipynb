## Pubmed citation graph data retrieval

I scraped the pubmed database for (A) BMI faculty papers, (B) papers that cite these publications, and (C) papers that were cited by these publications to generate a BMI publication network. I limited the faculty papers to at most 250 papers per faculty but even that resulted in >250000 "first-degree" (B+C) citations!


```python
### get BMI faculty articles
faculty = pd.read_csv("BMI_faculty.csv", header=None)

# construct entrez query
url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&retmax=250&retmode=json&term="
save_faculty_articles = {}

for author in faculty[0]:
    print(author)

    id_response = requests.post(url+author)
    save_faculty_articles[author] = id_response.json()['esearchresult']['idlist']

# save faculty articles (I save at every step because I am paranoid and also it has actually been helpful not to be blocked by the Entrez API)
with open('faculty_data.txt', 'w') as outfile:
    json.dump(save_faculty_articles, outfile)
    
    
```

```python
### get all articles that cite BMI faculty articles
with open('faculty_data.txt', 'r') as file:
    faculty_articles = json.load(file)

# construct entrez query 
url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/elink.fcgi?dbfrom=pubmed&linkname=pubmed_pubmed_citedin&retmode=json&id="

for author in faculty_articles.keys():
    print(author)
    ids = faculty_articles[author]
    id_response = requests.post(url+",".join(ids))
    cited_by = id_response.json()['linksets'][0]["linksetdbs"][0]["links"]
    
    faculty_articles[author] = [ids, cited_by]

# save faculty articles 
with open('faculty_data_citedby.txt', 'w') as outfile:
    json.dump(faculty_articles, outfile)
    
    
    
```

```python
### get all articles that BMI faculty articles cite
with open('faculty_data_citedby.txt', 'r') as file:
    faculty_articles = json.load(file)

# construct entrez query 
url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/elink.fcgi?dbfrom=pubmed&linkname=pubmed_pubmed_refs&retmode=json&id="

for author in faculty_articles.keys():
    print(author)
    ids = faculty_articles[author]
    id_response = requests.post(url+",".join(ids[0]))
    
    cited_in = id_response.json()['linksets'][0]["linksetdbs"][0]["links"]
    ids.append(cited_in)
    
    faculty_articles[author] = ids #faculty articles; cite faculty articles, cited in faculty articles

# save faculty articles
with open('faculty_citation_network.txt', 'w') as outfile:
    json.dump(faculty_articles, outfile)

    
```

## Building the BMI faculty citation graph

I then generated a heterogenous graph with faculty and the first-degree citations as nodes. I limited this graph to only the first 100 citing articles and 100 cited articles for each faculty member. Surprisingly, this created a mostly connected network


```python
### generate BMI faculty citation network

# open file
with open('faculty_citation_network.txt', 'r') as file:
    faculty_articles = json.load(file)

# create graph
G = nx.DiGraph()

for author in faculty_articles.keys():
    #G.add_nodes_from(faculty_articles[author][0])
    G.add_nodes_from(faculty_articles[author][1][:100]) # only take first 100 
    G.add_nodes_from(faculty_articles[author][2][:100])
    G.add_node(author)
    
    # A --- is cited by ---> B
    G.add_edges_from([(author, node) for node in faculty_articles[author][1] if node in G.nodes]) # papers citing BMI faculty
    G.add_edges_from([(node, author) for node in faculty_articles[author][2] if node in G.nodes]) # cited in BMI faculty papers

# remove nodes with 0 in-degree (impossible to reach)
remove_nodes = []
for node in G.in_degree:
    if node[1] < 1:
        remove_nodes.append(node[0])
        
G.remove_nodes_from(remove_nodes)

# save to file
nx.write_adjlist(G, "faculty_citation_network.adjlist", delimiter=";")

### generate a mini-graph
faculty = ["Michael Keiser", "Atul Butte", "Martin Kampmann", "Steven Altschuler", "Lani Wu"]
sub_nodes = []

for a in faculty:
    for b in faculty:
        path = nx.shortest_path(G, source=a, target=b, weight=None)
        sub_nodes.extend(path)
        
# get a tiny graph
G_sub = nx.subgraph(G, sub_nodes)

# save to file
nx.write_adjlist(G_sub, "tiny_citation_network.adjlist", delimiter=";")


    
```

```python
### Depth-first traversal 

G = nx.read_adjlist("tiny_citation_network.adjlist", create_using=nx.DiGraph, delimiter=";")
visited = []

def dfs_traversal(start):
    visited.append(start)

    for neighbor in G.neighbors(start):
        if neighbor not in visited:
            dfs_traversal(neighbor)
    
dfs_traversal("Michael Keiser")
visited
    
```

```python
### Depth-first search  

G = nx.read_adjlist("tiny_citation_network.adjlist", create_using=nx.DiGraph, delimiter=";")
visited = []

def dfs_traversal(start, end):
    visited.append(start)
    
    if start==end:
        return visited

    for neighbor in G.neighbors(start):
        if neighbor not in visited:
            path = dfs_traversal(neighbor, end)
            if path: return path
    
dfs_traversal("Michael Keiser","Atul Butte")

```
